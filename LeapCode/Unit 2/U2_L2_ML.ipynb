## Lesson 2.2: What is Logistic Regression?

Logistic Regression is used to predict **yes/no** or **true/false** answers.

Linear Regression uses a straight line to fit all the points, but Logistic Regression draws a curve (called a sigmoid curve) that helps it decide between “yes” or “no”

<img src="../Photos/DecisionCurve.png" width="500">

Anything below the decision bound is 0 (no/false) while anything above  is 1 (yes/true)

### Code Demo (Python)

Let's say we have some data that looks like this

Student 1: 40% on the test, didn't pass

Student 2: 50% on the test, didn't pass

Student 3: 60% on the test, didn't pass

Student 1: 70% on the test, passed

Student 2: 80% on the test, passed

Student 3: 90% on the test, passed

We can use logistic regression to see if a student who scored 66% will pass

````python
from sklearn.linear_model import LogisticRegression
import numpy as np

X = np.array([[40], [50], [60], [70], [80], [90]])
y = np.array([0, 0, 0, 1, 1, 1])

model = LogisticRegression()
model.fit(X, y)

# Predict if someone with a score of 65 will pass
print("Prediction for score of 66:", model.predict([[66]]))
````

Here's what's happening step-by-step:

1. **Importing the Model**:
   
   ```python
   from sklearn.linear_model import LogisticRegression

We can use the same library from the last lesson, Sci-kit learn, to import the Logistic Regression model

2. **Creating the Data**:

   ```python
    X = np.array([[40], [50], [60], [70], [80], [90]])
    y = np.array([0, 0, 0, 1, 1, 1])

X is our input which in this case is the test score

y is our output which in this case is if they failed or not

The code creates a table like the one shown below

3. **Training the Model**:
   
   ```python

    model.fit(X, y)

We define the model as the Logistic Regression model. We then train the model using .fit on the data we made earlier. 

.fit is a function that creates the sigmoid curve that matches the datapoints we gave the model. Here is what the line would look like

<img src="../Photos/DataPointLogReg.png" width="600">

4. **Making a Prediction**:

    ```python
    print("Prediction for score of 66:", model.predict([[66]]))


Finally, the model then uses the sigmoid curve to predict what the value of passed (y) will be when the score (X) is 66. The model used the curve and predicted that when the test score is 66, the y value will be 0.637688. Since the point is above the decision bound, the model predicts that the students passed (y=1).
