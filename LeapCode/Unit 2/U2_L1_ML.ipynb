## Lesson 2.1: What is Linear Regression?


Let's say you want to predict the test score a student will get based on how many hours they study. 

Here we have some data about students who studied different amounts and what test score they got. 

Student 1: studied for 1 hour and got a test score of 78%

Student 2: studied for 2.5 hours and got a test score of 89%

Student 3: studied for 4 hours and got a test score of 98%

We can use this data to predict what a student's test score will be based on how much they studied using **Linear Regression**

Linear Regression reads data points and creates a line of best fit for the data. The line of best fit may not cover every data point, but it can give us a pretty good prediction.

<img src="../Photos/LinearRegression.png" width="350">

### Code Demo (Python)

````python
# We import scikit-learn's Linear Regression model. Scikit-learn is a popular library for machine learning
from sklearn.linear_model import LinearRegression
import numpy as np

# We place the data earlier into an array
X = np.array([[1], [2.5], [4]])
y = np.array([78, 89, 98])

# Defining the model
model = LinearRegression()
model.fit(X, y)

print("Prediction for 3.5:", model.predict([[3.5]]))
````

In this simple example, we demonstrate the fundamental idea of models learning from data to make predictions.

Here's what's happening step-by-step:

1. **Importing the Model**:
   
   ```python
   from sklearn.linear_model import LinearRegression

We define the model as the Linear Regression model from sci-kit learn. We then train the model using .fit on the data we made earlier.
