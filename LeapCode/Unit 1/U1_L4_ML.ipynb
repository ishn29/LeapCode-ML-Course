## Lesson 1.4: What is a Dataset?

Before we teach a computer to learn, we need to give it something to learn **from**, that's where **datasets** come in

---

### What is a Dataset?

A **dataset** is a **table of data**, kind of like a spreadsheet in Excel or Google Sheets.

Each **row** is one item (like a person, animal, or object).  
Each **column** is a detail about that item (like age, color, or size).

---

#### Example Dataset: Student Info

| Name   | Age | Height | Passed? |
|--------|-----|--------|---------|
| Lily   | 13  | 150    | Yes     |
| Max    | 14  | 160    | No      |
| Aiden  | 12  | 145    | Yes     |
| Sophia | 13  | 155    | Yes     |

- Each row is a student.
- Each column is a **feature** (something we know about them).
- The last column, "Passed?", is the **target** (the thing we want to predict).

---

### What are Features and Target?

- **Features** = the input columns (things we give to the computer)
  - Examples: Age, Height
- **Target** = the answer column (what we want the computer to learn to predict)
  - Example: Passed?
