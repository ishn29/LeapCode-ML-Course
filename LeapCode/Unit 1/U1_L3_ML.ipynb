## Lesson 1.3: Deep Learning and Neural Networks

### What is Deep Learning?

**Deep learning** is a special kind of machine learning that uses **layers of neurons** to learn from lots and lots of data.

Think of it like an artificial brain made of **math and patterns**.

> Machine Learning = learning simple patterns  
> Deep Learning = learning really complicated stuff (like faces, voices, or language)

---

### What is a Neural Network?

A **neural network** is the main tool used in deep learning. It’s made of tiny parts called **neurons**.

Each **neuron**:
- Takes some numbers as input
- Does a simple calculation
- Sends the result to the next neuron

A **neural network** is like a team of neurons working together:

### 1. Input Layer

- This is where the network **receives data**.
- Each **input** is a number (like a pixel value or a feature).
- Example: If you're trying to guess someone's height, the inputs might be:
  - Age = 14
  - Shoe size = 8

> The input layer is like your **senses** (eyes, ears), it collects information.

---

### 2. Hidden Layers

- These layers do the **thinking** behind the scenes.
- Each neuron looks at the inputs, does math, and sends results to the next layer.
- There can be **1 or many** hidden layers.

> The hidden layers are like your **brain** — processing what your eyes and ears saw.

---

#### What happens inside a hidden layer?
1. Each neuron gets numbers from the previous layer.
2. It multiplies and adds them together.
3. It passes the result through an **activation function** (like a decision gate).
4. The output goes to the next layer.

---

### 3. Output Layer

- This is where the network gives its **final answer**.
- It could be:
  - A number (e.g. predicted height)
  - A class (e.g. "cat" or "dog")
  - A probability (e.g. 80% chance it’s a cat)
