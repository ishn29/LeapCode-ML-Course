## Lesson 1.2: Unsupervised Learning

**Unsupervised learning** is when we give the computer data **without any answers**, and ask it to find patterns or groups all by itself.

It’s like giving the computer a puzzle and not telling the computer what the puzzle will make

---

### Example: Sorting Snacks

Imagine you have a bunch of snacks:

| Snack Name | Color | Shape     | Crunchy? |
|------------|-------|-----------|----------|
| Chips      | Yellow| Round     | Yes      |
| Cookies    | Brown | Round     | Yes      |
| Carrots    | Orange| Stick     | Yes      |
| Marshmallow| White | Fluffy    | No       |
| Apple      | Red   | Round     | No       |

You give this to the computer and say:

> “Sort these snacks into groups based on how similar they are.”

There are no labels like “Group A” or “Group B.” The computer might figure out:

- Group 1: Crunchy snacks
- Group 2: Not crunchy snacks

Or maybe:

- Group 1: Round shapes
- Group 2: Stick/Fluffy shapes

---

### Why is it called "Unsupervised"?

Because we don’t give the computer any answers!

- No labels
- Just raw data

The computer has to **figure things out on its own**, like grouping, spotting patterns, or finding outliers (weird items).

---

#### Real-World Examples:
- Grouping customers by shopping habits
- Finding patterns in music, pictures, or behavior
- Detecting unusual data (like fraud or fake reviews)
