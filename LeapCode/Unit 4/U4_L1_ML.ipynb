## Lesson 4.1: Overfitting and Underfitting

## What is Overfitting and Underfitting?

### Underfitting:

- The model is **too simple**
- It **misses patterns** in the data
- Performs badly on training and test data

Example: Trying to fit a straight line to wiggly data

---

### Overfitting:

- The model is **too complex**
- It fits **every detail**, even the noise
- Works well on training data, but **bad on new data**

Example: Memorizing the homework instead of learning how to solve it

---

### Good Fit:

- Learns the true pattern
- Performs well on both training and test data

---

## Visual Example

Let's generate some fake data to show this:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Make fake data
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.2, X.shape[0])

# Plot data
plt.scatter(X, y, label="Data", color="black")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Training Data (Noisy Sin Wave)")
plt.show()
```

### Trying Three Models

Let’s train three models:

Degree 1 = Underfitting

Degree 15 = Overfitting

Degree 4 = Just Right

```python
degrees = [1, 4, 15]
colors = ['blue', 'green', 'red']

plt.scatter(X, y, label="Data", color="black")

for d, color in zip(degrees, colors):
    model = make_pipeline(PolynomialFeatures(d), LinearRegression())
    model.fit(X, y)
    X_test = np.linspace(0, 5, 100).reshape(-1, 1)
    y_pred = model.predict(X_test)
    plt.plot(X_test, y_pred, label=f"Degree {d}", color=color)

plt.legend()
plt.title("Underfitting vs. Overfitting")
plt.show()
```

---
### What You Should See


Blue Line (Degree 1): Too simple — underfits

Red Line (Degree 15): Follows noise — overfits

Green Line (Degree 4): Smooth and accurate — just right

### Why This Matters in Real Life


A model that overfits will look perfect in training, but fail in the real world

A model that underfits will never make useful predictions

We want the model to generalize — to make good predictions on new, unseen data
