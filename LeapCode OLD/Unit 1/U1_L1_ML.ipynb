## Lesson 1.1: Supervised Learning


### What is Machine Learning?

**Machine Learning (ML)** is a subset of artificial intelligence (AI) that enables systems to learn patterns from data and make predictions or decisions without being explicitly programmed for every task.


#### Traditional Programming vs Machine Learning
| Traditional Programming | Machine Learning |
|-------------------------|------------------|
| Input + Rules → Output  | Input + Output → Model (Learns Rules) |

### What is Supervised Learning?

Supervised learning is when we teach a computer using data that already has **both the input and the correct answer**.

It's like giving the computer a bunch of flashcards to study:
- One side of the card: a question (input)
- Other side: the correct answer (label)

---

### Example: Study Time and Test Scores

Imagine you're a teacher and you have this data:

| Hours Studied | Test Score |
|---------------|------------|
| 1             | 50         |
| 2             | 60         |
| 3             | 70         |
| 4             | 80         |

- The **input** is: Hours Studied  
- The **output (label)** is: Test Score

You give this data to the computer and say:

> “Learn the pattern between how much a student studies and what score they get.”

Now, if we ask the computer:

> **“If a student studies for 5 hours, what score will they get?”**

It might answer:

> **90**

Because it learned from the earlier examples that had the test score (label). This is a simple example of supervised learning

---

### Why is it called "Supervised"?

Just like a teacher **supervises** a student, you're guiding the computer by giving it the correct answers while it learns.

- You give it examples.
- You give it answers.
- It learns to make predictions on new data.
